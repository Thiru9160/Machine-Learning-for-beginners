```python
#plot a graph  with dataframe
#dataframe
data={
    "Cricket_bat":['SG','KOOKABURRA','ss','BDM','spatan'],
    'price':[2000,2200,2400,2600,2800],
    'Weight in grams':[1100,1200,1250,1300,1350]
}
#dataframe
dataframe=pd.DataFrame(data)
#plot a line using plt() method
#here x and y points are nothing but columns in dataframe
plt.plot('price','Weight in grams')

#adding the titles and position
#title()--set the title to tht graph
plt.title('Employee',loc='right')#by default the position in center

#adding the labels to xpoints and ypoints
plt.xlabel("Age of employee")
plt.ylabel("salary of employee")

#adding grid lines
plt.grid()

#display the figure
plt.show()
```

**BAR GRAPH**

It is pictorail reprsentation using rectangular bars of different height.These bars reprsents the value of data.

bar() method--used to plot bar graph.

```python
student=np.array(['A','B','C','D','E'])
marks=np.array([20,40,80,90,75])
colors=['red','orange','green','yellow','purple']
#plot a bar graph
#here,X and Y co-ordinates are student and marks
plt.bar(student,marks,color=colors)
#label the x and y co-ordinates.
plt.xlabel('Student Name')
plt.ylabel('Marks')
#make a title
plt.title("Student Marks")
#show the bar graph.
plt.show()
```

**PIE CHART**

A pie chart displays data in a pictorial form, divided into slices(parts).

pie() method--to plot pie chart

First parameter: X value

second parameter: label parameter for each X value

Third parameter : %


```python
crickter=np.array(['virat','Rohit','Pant','Iyer','KLRahul'])
runs=np.array([85,99,55,65,45])
colors=['gold','lightgreen','lightskyblue','red','purple']
#plo a pie chart --pie()
plt.pie(runs,labels=crickter,autopct='%1.2f%%',colors=colors)
#plot a title
plt.title("Runs scored in a cricket match")
#display a pie chart
plt.show()
```

**HISTOGRAM**

It is also graphical reprsentation to show frequency distributions

hist()--to plot a histogram.

```python
arr=np.array([10,54,36,45,68,32,8,3,24,55,66])#0-10(2),30-40(2),50-80(4)
#plot a histogram
#bins parameter sets the bins
colors=['gold','lightgreen','lightskyblue','red','purple']
plt.hist(arr,bins=[0,10,20,30,40,50,70,80],color='r')
#labels
plt.xlabel('marks')
plt.ylabel('student')
#title
plt.title("student marks")
#displat the histogram
plt.show()
```

**SCATTER PLOT**

It is used to find relationship between two variables. A dot represents each value inthe dataset.

scatter() method--used to plot scatter plot.

```python
team1_score=[25,37,34,38,27,40,9,18]
team2_score=[7,22,40,29,27,10,19,31]
#score range(lower order batsman) in cricket match
score_range=[5,10,15,20,25,30,35,40] # Corrected score_range to match the length of team scores
#plot a scatter
plt.scatter(team1_score,score_range,color='r')
plt.scatter(team2_score,score_range,color='b')
#labels
plt.xlabel('team score')
plt.ylabel('score range')
#add title
plt.title('cricket match')
#display scatter plot
plt.show()
```

**LINE GRAPH**

It is used to compare two varibales,display on the x and y axis.

Horizontal axis is called x-axis

vertical axis is called y-axis.

plot() method--used to plot line graph

```python
months=np.array(['jan','feb','march','april','may','june'])
rainfall=np.array([60,65,40,45,70,85])
#plot a line graph
plt.plot(months,rainfall,color='b')
#labels
plt.xlabel("months")
plt.ylabel("rainfall")
#title
plt.title("monthly rainfall")
#display the graph
plt.show()
```
