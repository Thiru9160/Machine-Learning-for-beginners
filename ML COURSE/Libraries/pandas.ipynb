**Pandas**

pandas is designed for data manipulation and analysis especially when dealing with structured data like csv files or sql tables.
There are two types:
1. pandas series
2.pandas dataframe

```python
#pandas series is like a column in a table.It is a 1-d array holding data of any type.
#importing libraries
import pandas as pd
import numpy as np
```

```python
#string
country=['India','U.S.A','England','Russia','Nepal']
print(country)
pd.Series(country)#pd.series()-->used to convert any dataype into pandas series.
```

```python
#integers
marks=[78,89,67,98,65]
print(pd.Series(marks))
#custom index
subjects=['Maths','English','Science','Python','Java']
pd.Series(marks,index=subjects)#index parameter is used to give our own index or custom index
#setting name
marks=pd.Series(marks,index=subjects,name="sam marks")#name is used togive name for pandas series table
marks
```

**Series from dictionary**

```python
marks={'maths':78,
       'python':87,
       'java':28,
       'english':80
       }#key-value pari
marks_series=pd.Series(marks,name='sidd marks')#key --> index, value --> value
marks_series
```

**Series Attributes(properties)**

```python
#size:Returns the number of elements in the series
print(marks_series.size)
#dtype:returns datatype of the value in series
print(marks_series.dtype)
#name:returns the name of the series
print(marks_series.name)
#is_unique: returns true if every value in series is unique else false.
print(marks_series.is_unique)
#index
print(marks_series.index)
#values
print(marks_series.values)
#type
type(marks_series.values)
```
